# Q.2. In which year highest number of the TV Shows & Movies were released ? Showing with Bar Graph.

### to_datetime

```python
data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce')
```

### dt.year.value_counts()

```python
# It counts the occurrence of all individual Years in Date column.
data['Date_N'].dt.year.value_counts()
```

### Bar Graph

```python
import matplotlib.pyplot as plt

top_years = data['Date_N'].dt.year.value_counts().head(5)
top_years.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Top 5 Years')
plt.show();
```
